## Customer segmentation (RFM technique)

```python
# Calculate Recency (days since last purchase)
max_date = clint_demo_clean['InvoiceDate'].max()
recency = clint_demo_clean.groupby('CustomerID')['InvoiceDate'].max().reset_index(name='LastPurchaseDate')
recency['Recency'] = (max_date - recency['LastPurchaseDate']).dt.days

# Calculate Frequency (unique InvoiceNo per CustomerID)
frequency = clint_demo_clean.groupby('CustomerID')['InvoiceNo'].nunique().reset_index(name='Frequency')

# Calculate Monetary (Quantity * UnitPrice)
clint_demo_clean['Monetary'] = clint_demo_clean['Quantity'] * clint_demo_clean['UnitPrice']

# Calculate Monetary (total spend per customer)
monetary = clint_demo_clean.groupby('CustomerID')['Monetary'].sum().reset_index(name='Monetary')

# Merge all RFM metrics
rfm = frequency.merge(recency[['CustomerID', 'Recency']], on='CustomerID')
rfm = rfm.merge(monetary, on='CustomerID')

# Change place of columns
rfm = rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary']]

# Use display for interactive output in notebooks
display(rfm)

# Show the first few rows
rfm.head()
```

```python
rfm['r_score'] = pd.cut(rfm['Recency'], 4, labels=[4, 3, 2, 1])
rfm['f_score'] = pd.cut(rfm['Frequency'], 4, labels=[1, 2, 3, 4])
rfm['m_score'] = pd.cut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])

rfm.sample(5)
```

```python
rfm['rfm_score'] = rfm['r_score'].astype(str) + rfm['f_score'].astype(str) + rfm['m_score'].astype(str)
rfm['rfm_score'] = rfm['rfm_score'].astype(int)
rfm.head()
```

```python
rfm['customer_segmentation'] = pd.cut(rfm['rfm_score'], 4, labels=['Bronze', 'Silver', 'Gold', 'Platinum'])
rfm.head()
```
